```python
# 1. Download the following dataset found at the following link: 
# https://healthdata.gov/dataset/certificate-need-applications-beginning-1974¶
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('Certificate_of_Need_Applications__Beginning_1974.csv')
dataset

```

```python
# 2. What type of license is this Dataset shared under?
# The dataset is shared under Open Database License (ODbL)
# http://opendefinition.org/licenses/odc-odbl/
# https://opendatacommons.org/licenses/odbl/summary/
```

```python
# 3. Omit all rows/observations with missing 'Total Project Cost' information
dataset = dataset[pd.notnull(dataset['Total Project Cost'])]
dataset
```

```python
# 4. Filter out the observations for the range of years 2014-2019
dataset = dataset[(pd.to_datetime(dataset['Project Status Date']) < pd.to_datetime('2014-01-01')) | (pd.to_datetime(dataset['Project Status Date']) > pd.to_datetime('2019-12-31'))]
dataset
```

```python
# 5. Develop an informative plot of total project cost by facility type
newDF = pd.read_csv('Certificate_of_Need_Applications__Beginning_1974.csv')

with sns.axes_style(style='ticks'):
   g = sns.barplot("Total Project Cost", "Facility Type",data = newDF)
```

```python
# 6. Stratify the plot you developed by year
newDF['Year'] = pd.DatetimeIndex(newDF['Project Status Date']).year
newDF
with sns.axes_style(style='ticks'):
   g = sns.barplot("Total Project Cost", "Facility Type", hue="Year",data = newDF)
```

```python
# 7. Develop an informative visual plot of your choice
with sns.axes_style(style='ticks'):
    g = sns.barplot("Year","Total Project Cost", data = newDF)
```
