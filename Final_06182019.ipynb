```python
# Exercise 1
# a) Import the posted Excel file, Washington_Health_Workforce_Survey_Data.xls
import pandas as pd
data = pd.read_excel('Washington_Health_Workforce_Survey_Data.xls')
data
```

```python
# Exercise 1
# b) Calculate and interpret a correlation measure for 'NumberYearsPracticeWashington' 
# and 'BirthYear' variables, what does the result indicate?
data[["NumberYearsPracticeWashington","BirthYear"]].corr()
# As the result shown below, BirthYear and NumberYearsPracticeWashington are negatively correlated 
# and their correlation is -0.587643. Which means about 59 percent of the time, BirthYear will 
# decrease when NumberYearsPracticeWashington increase
```

```python
# Exercise 1
# c) Impute the missing values for the 'AnnualWeeksWorked' variable by the median.
median = data["AnnualWeeksWorked"].median()
data["AnnualWeeksWorked"].fillna(median, inplace=True)
data
```

```python
# Exercise 1
# d) Calculate the mean for the 'AnnualWeeksWorked' variable.
data["AnnualWeeksWorked"].mean()
```

```python
# Exercise 1
# e) Compute the same staistics (mean and median) of 'AnnualWeeksWorked' variable' for male and females seperately.
Male_filter=data["Sex"] == 'Male'
Female_filter=data["Sex"] == "Female"
print('Male - median: ', data["AnnualWeeksWorked"][Male_filter].median())
print('Male - mean: ', data["AnnualWeeksWorked"][Male_filter].mean())
print('Female - median: ', data["AnnualWeeksWorked"][Female_filter].median())
print('Female - mean: ', data["AnnualWeeksWorked"][Female_filter].mean())
```

```python
# Exercise 2
# Develop and code a word parsing program which repalces certain grammatical elements in a phrase with user 
# input as follows:
# The octopus snagged the blue fish
# The program will then ask the user to enter a new word for subject ( octopus), adjective (blue) and object (fish), 
# let us suppose the user enters subject=shark, verb= ate, adjective=huge, and object=tuna, the output will be as follows:
# The shark bit the huge tunaÂ¶

print("Please enter an subject, an adjective and an object to replace the sentence: ")
print("The octopus(subject) snagged the blue(adjective) fish(object)")
original = "The octopus snagged the blue fish"
subject = input(str('subject: '))
adjective = input(str('adjective: '))
object = input(str('object: '))
result = original.replace("octopus", subject)
result = result.replace("blue", adjective)
result = result.replace("fish", object)
print("Your sentense is: ", result)
```

```python
# Exercise 3:
# Develop and code a guessing program, which asks the user to generate a random number in the range (1:100). 
# The program is not being informed of the value of the randomly generated number so that it can guess the number. 
# Once you have generated the number develop your guessing algorithm as follows: the program will begin guessing by 
# selecting a number in the middle of the range. The program will present this number to the user, the user will 
# then indicate whether the number is too high or too low. If it is too high then the program will present a 
# number that is one integer lower, and if it too low then the program will present a number that is 1 integer higher.
# For example if the program guesses 54 and the user indicates too high then the next number the program will 
# present is 53.
num = int(input("Please enter an integer between 1 and 100: "))
guess = 50
while True:
    correct = input('Is the number ' + str(guess) + '? (please enter "T" or "F")')
    if correct == "T":
        break
    else:
        hiLow = str(input('Is it higher than ' + str(guess) + ' or lower than ' + str(guess) + '? (please enter "higher" or "lower")'))
        if hiLow == "higher":
            guess = guess + 1
        else:
            guess = guess - 1
print('OOh ha! Your number is: ', guess)
            
```
